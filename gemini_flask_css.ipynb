### <center><u><span style="color:deepskyblue">CSS</span> for the <span style="color:green">Flask</span> App</u></center>
#### <i>Why do this? I mean, who cares??</i>
#### <u>Couple of reasons</u>:
> ##### 1.) I couldn't get a consistent-looking <b><span style="color:red">PDF</span></b> tailored resume output. Sometimes the font differed, spacing was off - if you were a Recruiter looking at the resume produced by one particular run, you'd have thought I was a clown in big red shoes applying for a Data Engineering position;
> ##### 2.) Because we're exporting the PDF using <b><span style="color:green">WeasyPrint</span></b>, <b><span style="color:green">WeasyPrint</span></b> will inevitably go through some changes. By explicitly defining the style of the resume we want, we can encode it into our function so that the output stays consistent.

#### <span style="color:deepskyblue">CSS</span> is a styling language (not really a "programming language") you use to tell your computer how you want things to look: layouts, font types, colors, etc.
#### You can get pretty creative with <span style="color:deepskyblue">CSS</span>, but <span style="color:coral">Applicant Tracking Systems</span> (<span style="color:coral">ATS</span>) don't really react will to highly creative styling.
#### That's why the <span style="color:deepskyblue">CSS</span> below is designed to give us a minimialistic yet professional-looking <span style="color:red">PDF</span> at the end.
#### <u>Explanation of <span style="color:green">Imports</span></u>:
> ##### <u><b><span style="color:deepskyblue">markdown</span></b></u>: we're adusting the resume using markdown, and after the adjustments, we're exporting it to a <b><span style="color:red">PDF</span></b>;
> ##### <u><b><span style="color:blue">IPython</span></b></u>: so we can see the enriched formatting inside this Jupyter Notebook; and
> ##### <u><b><span style="color:green">WeasyPrint</span></b></u>: we're using this as our conversion tool in the program
> ##### <u><b><span style="color:red">Aliases</span></b></u>: there to avoid confusion

```python
from markdown import markdown as markdown_to_html
from IPython.display import display, HTML
from weasyprint import HTML as WeasyHTML 
```

```python
markdown_path = "/Users/nicholasjoseph/Desktop/nj_safe_master_resume.md"
```

```python
# Read the resume Markdown:
markdown_text = ""
try:
    with open(markdown_path, "r", encoding="utf-8") as file:
        markdown_text = file.read()
except FileNotFoundError:
    print(f"Ruh-roh. Markdown file '{markdown_path}' not found. Please make sure you've got the right path in 'markdown_path.'")
except Exception as e:
    print(f"Error reading file: {e}.")
    markdown_text = None
```

```python
# show the raw markdown
if markdown_text:
    basic_html_content = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <title>Original Resume Content (Unstyled Markdown)</title>
    </head>
    <body>
        <h1>Original Resume Content (Unstyled Markdown)</h1>
        <pre style="white-space: pre-wrap; word-wrap: break-word; background-color: #f4f4f4; border: 1px solid #ddd; padding: 15px; border-radius: 5px; font-family: 'Courier New', monospace;">{markdown_text}</pre>
    </body>
    </html>
    """
    print("--- BEFORE: Unstyled HTML (Raw Markdown Content) ---")
    display(HTML(basic_html_content))
else:
    print("The Markdown wasn't loaded, so we can't show you the super cool HTML.") 
```

```python
# save this for later:
raw_resume_markdown = markdown_text
```

#### Now to define some <span style="color:deepskyblue">CSS</span>. 
#### All but the <span style="color:darkmagenta">@page</span> are visible in the browser display. <span style="color:darkmagenta">@page</span> formats like you would in Pages or Word - "How big is the paper? What are the margins?" That kind of stuff.
> ##### 1.) <b><span style="color:darkmagenta">@page</span></b> - setting standard margins
> ##### 2.) <b><span style="color:firebrick">body</span></b> - setting font style, size, and color
> ##### 3.) <b><span style="color:firebrick">h1, h2, h3</span></b> - how much we want the headers (sections) to stand out
> ##### 4.) <b><span style="color:firebrick">p</span></b> - what we want the paragraph text (what you did at your job) to look like (spacing for readability)
> ##### 5.) <b><span style="color:firebrick">ul</span></b> - sets up spacing for your bullet points - your lists.

```python
css_styles = """
    @page {
        margin: 1in;
        }
        
    body {
        font-family: 'Times New Roman', sans-serif;
        font-size: 11pt;
        line-height: 1.5;
        color: #222;
        padding: 20px;
        }

    h1, h2, h3 {
        font-weight: bold;
        margin-bottom: 10px;
        }
        
    p {
        margin: 0 0 12px;
        }
        
    ul {
        padding-left: 20px;
        }
        """
```

#### Converting that earlier markdown resume (that '<b>raw_resume_markdown</b>' variable) by giving the <span style="color:deepskyblue">CSS</span> some <span style="color:darkorange">HTML</span> to work on.

```python
css_styled_resume = markdown_to_html(raw_resume_markdown)

styled_html_content = f"""
<!DOCTYPE html>
<html>
<head>
    <title>Styled Resume Content:</title>
    <style>
        {css_styles}
    </style>
</head>
<body>
    <h1>Styled Resume Content (Rendered from Markdown)</h1>
    {css_styled_resume}
</body>
</html>
"""
```

#### <b><i>Alright, Bucko. Let's see it:</b></i> 

```python
print("\n--- AFTER: The Styled HTML (Conceptual Layout) ---")
display(HTML(styled_html_content))
```

#### Even though it <i>looks</i> like a <span style="color:red">PDF</span>, it's not. We still have to convert it, and that's where <span style="color:green">WeasyPrint</span> comes in.

```python
WeasyHTML(string=styled_html_content).write_pdf("styled_resume.pdf")
print("\n'styled_resume.pdf' has been generated in your notebook directory.")

```

#### 'styled_resume.pdf' is within this particular directory to show inquiring minds how it all worked out.
