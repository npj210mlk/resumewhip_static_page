### <b><u>Why <span style="color:darkorchid">Stripe</u></span>?</b>
> #### 1.) As much as I want to make this service free, APIs and Server Instances (in this case, Render) are expensive;
> #### 2.) Therefore, we have to authenticate users with something other than email addresses to prevent abuse; and
> #### 3.) <span style="color:darkorchid">Stripe</span> is a trusted source for doing all this, using payment methods for authentication instead of emails.

#### <b>N.B.:
</b> <span style="color:darkorchid">Stripe</span> needs to be installed either in the current directory (using '<b><span style="color:blue ; font-family:menlo">pip install stripe</b></span>') or inside / in-line with the current Jupyter Notebook, using '<b><span style="color:green ; font-family:menlo">!pip install stripe</b></span>.' We've elected to do it using the first option, with the accessory information stored inside a '.env' file.
